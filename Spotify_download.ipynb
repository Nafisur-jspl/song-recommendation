```python
import pandas as pd
```

```python
songs = pd.read_csv('song_list.csv')
```

```python
sample_songs = songs.head(10)
```

```python
sample_songs = sample_songs.song.tolist()
```

```python
sample_songs
```

```python
import spotipy
spotify = spotipy.Spotify()
```

```python
results = []
for song in sample_songs:
    result = spotify.search(q='track:' + song, type='track')
    results.append(result)
```

```python
len(results)
```

```python
print len(results[0]['tracks']['items'])
print results[0]['tracks']['items'][0]['name']
print results[0]['tracks']['items'][0]['popularity']
print results[0]['tracks']['items'][0]['duration_ms']
```

```python
for result in results:
    for item in result['tracks']['items']:
        print item['name'], item['popularity'], item['duration_ms']
```
